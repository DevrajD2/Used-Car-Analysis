```python
avg_price_by_age = df.groupby("Age")["AskPrice"].mean().reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_price_by_age, x="Age", y="AskPrice", palette="mako")
plt.title("Average Ask Price vs Car Age")
plt.xlabel("Car Age (Years)")
plt.ylabel("Average Ask Price (in thousands â‚¹)")
plt.show()
```

```python
avg_price_by_fuel = df.groupby("FuelType")["AskPrice"].mean().reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_price_by_fuel, x="FuelType", y="AskPrice", marker='o', linewidth=2, color='green')
plt.title("Average Ask Price by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Average Ask Price (in thousands â‚¹)")
plt.grid(True)
plt.show()
```
